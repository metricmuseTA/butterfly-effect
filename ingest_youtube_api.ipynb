```python
# — Ingest YouTube trailer stats for each film
from googleapiclient.discovery import build
import pandas as pd

youtube = build('youtube', 'v3', developerKey=YOUTUBE_API_KEY)

results = []

for film in FILM_LIST:
    title = film["title"]
    
    try:
        # Search for official trailer
        search_response = youtube.search().list(
            q=f"{title} official trailer",
            part='id,snippet',
            maxResults=1,
            type='video'
        ).execute()
        
        if not search_response['items']:
            print(f"⚠️ {title} — no trailer found")
            continue
            
        video_id = search_response['items'][0]['id']['videoId']
        snippet = search_response['items'][0]['snippet']
        
        # Get video statistics
        stats_response = youtube.videos().list(
            part='statistics,snippet',
            id=video_id
        ).execute()
        
        if not stats_response['items']:
            print(f"⚠️ {title} — no stats found")
            continue
            
        stats = stats_response['items'][0]['statistics']
        
        results.append({
            'film':          title,
            'release_date':  film['release_date'],
            'genre':         film['genre'],
            'video_id':      video_id,
            'trailer_title': snippet['title'],
            'published_at':  snippet['publishedAt'],
            'view_count':    int(stats.get('viewCount', 0)),
            'like_count':    int(stats.get('likeCount', 0)),
            'comment_count': int(stats.get('commentCount', 0)),
        })
        
        print(f"✅ {title} — {int(stats.get('viewCount',0)):,} views")
        
    except Exception as e:
        print(f"❌ {title} — error: {e}")

# Combine results
if results:
    df_youtube = pd.DataFrame(results)
    print(f"\nTotal films captured: {len(df_youtube)}")
    display(df_youtube)
else:
    print("No data collected.")
```
